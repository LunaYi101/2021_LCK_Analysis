```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 전체 플레이어 데이터에서 원딜 데이터 추출
player_data = pd.read_csv('LCK 2021 Spring - Player Stats - OraclesElixir.csv')
player_data = player_data[player_data['Pos']=='Support']

# 편의를 위해 부수적인 기호 삭제
for col in player_data.columns:
    if isinstance(player_data[col].iloc[0], str):
        player_data[col] = player_data[col].map(lambda x: x.replace('%', ''))

# 전부 소수(float) 형으로 변경
for col in player_data.columns[3:]:
    player_data[col] = pd.to_numeric(player_data[col], downcast = 'float')
    
# K, D, A 평균으로 변경
for col in ['K', 'D', 'A']:
    player_data[col] = player_data[col]/player_data['GP']
        
player_data
```

```python
player_data['DMG%/KP'] = player_data['DMG%']
player_data['DMG%/KP'] /= player_data['KP']
player_data = player_data.sort_values('DMG%/KP', ascending=False)
player_data
```

```python
spring_rank = ['DWG KIA', 'Gen.G', 'Hanwha Life Esports', 'T1', 'DRX', 'Nongshim RedForce', 'KT Rolster', 'Afreeca Freecs', 'Liiv SANDBOX', 'Fredit BRION']

player_data.Team = pd.Categorical(player_data.Team, categories=spring_rank, ordered = True)
player_data.sort_values('Team', inplace=True)

spring_rank_list = [spring_rank.index(x)+1 for x in player_data.Team]
player_data['Spring_Rank'] = spring_rank_list

player_data
```

```python
# 스프링 시즌 등수와 상관 계수가 0.8이상인 지표 (3점)
rank_1_columns = player_data.corr()[player_data.corr()['Spring_Rank']<=-0.8].index
rank_1_columns = [item for item in rank_1_columns]

# 스프링 시즌 등수와 상관 계수가 0.6이상인 지표 (2점)
rank_2_columns = player_data.corr()[player_data.corr()['Spring_Rank']<=-0.6].index
rank_2_columns = [item for item in rank_2_columns if item not in rank_1_columns]

# 스프링 시즌 등수와 상관 계수가 0.4이상인 지표 (1점)
rank_3_columns = player_data.corr()[player_data.corr()['Spring_Rank']<=-0.4].index
rank_3_columns = [item for item in rank_3_columns if item not in rank_2_columns]
rank_3_columns = [item for item in rank_3_columns if item not in rank_1_columns]

# 선수별 점수 저장 DataFrame 만들기
tmp_list = np.array([0] * player_data['Player'].count())
ADC_score = DataFrame(data=tmp_list, index = player_data.Player, columns = ['Score'])

print(rank_1_columns)
print(rank_2_columns)
print(rank_3_columns)
```

```python
copied_player_data = player_data
for column in rank_1_columns:
    copied_player_data.sort_values(column, inplace=True) # 해당 지표로 정렬 역순
    copied_player_data['Rank'] = np.arange(copied_player_data['Player'].count()) # 순위 매기기
    ADC_score = ADC_score.add(DataFrame(data=Series(list(copied_player_data['Rank']*3), index=list(copied_player_data['Player'])),index = player_data.Player, columns = ['Score']))
```

```python
for column in rank_2_columns:
    copied_player_data.sort_values(column, inplace=True) # 해당 지표로 정렬 역순
    copied_player_data['Rank'] = np.arange(copied_player_data['Player'].count()) # 순위 매기기
    ADC_score = ADC_score.add(DataFrame(data=Series(list(copied_player_data['Rank']*2), index=list(copied_player_data['Player'])),index = player_data.Player, columns = ['Score']))
```

```python
for column in rank_3_columns:
    copied_player_data.sort_values(column, inplace=True) # 해당 지표로 정렬 역순
    copied_player_data['Rank'] = np.arange(copied_player_data['Player'].count()) # 순위 매기기
    ADC_score = ADC_score.add(DataFrame(data=Series(list(copied_player_data['Rank']), index=list(copied_player_data['Player'])),index = player_data.Player, columns = ['Score']))
```
